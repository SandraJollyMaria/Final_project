Split the data into Train and Test

```python
#The below code is used for the preprocessing of data and stratified splitting of a dataset.
#This code cleans up the dataset by removing rows that contain missing ratings and converts these ratings into integers.
#Afterward, it will perform a stratified split on the training and test datasets to preserve the distribution of ratings in both sets.
#This step ensures that all classes are included within each split.
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
print("Before : {}".format(len(data)))
dataAfter = data.dropna(subset=["reviews.rating"])
# Removes all NAN in reviews.rating
print("After  : {}".format(len(dataAfter)))
dataAfter["reviews.rating"] = dataAfter["reviews.rating"].astype(int)

split = StratifiedShuffleSplit(n_splits=10, test_size=0.2)
for train_index, test_index in split.split(dataAfter,
                                           dataAfter["reviews.rating"]):
    strat_train = dataAfter.reindex(train_index)
    strat_test = dataAfter.reindex(test_index)
```

```python
#The above code snippet is used to check that stratified splitting has worked correctly by showing the size of each dataset and making sure the test set still contains the same distribution of ratings as the original dataset.
print(len(strat_train))
print(len(strat_test))
round((strat_test["reviews.rating"].value_counts()* 100/len(strat_test)),2)
```

Sentiment Analysis

```python
def sentiments(rating):
    # Check if the rating is 5 or 4 and return "Positive"
    if (rating == 5) or (rating == 4):
        return "Positive"
    # Check if the rating is 3 and return "Neutral"
    elif rating == 3:
        return "Neutral"
    # Check if the rating is 2 or 1 and return "Negative"
    elif (rating == 2) or (rating == 1):
        return "Negative"
    # If the rating does not match any of the above, return "Unknown" (added for completeness)
    else:
        return "Unknown"
# Add sentiments to the data
strat_train["Sentiment"] = strat_train["reviews.rating"].apply(sentiments)
strat_test["Sentiment"] = strat_test["reviews.rating"].apply(sentiments)
print(strat_train["Sentiment"][:15])
```

```python
#The above code defines a function that generates a detailed report of a DataFrame's descriptive statistics and missing values, and saves it as an HTML file.
#The report includes statistical summaries for all columns and counts of missing data. The function is executed with the DataFrame, and a confirmation message is printed upon successful saving of the report.
#This automation facilitates easy sharing and reviewing of data summaries.
```

```python
#Automated Reporting
import pandas as pd

# Function to generate automated reports
def generate_report(df, filename='report.html'):
    report = df.describe(include='all').transpose()
    report['missing_values'] = df.isnull().sum()
    report.to_html(filename)
    print(f'Report saved to {filename}')

# Assuming df is your DataFrame with Amazon reviews
generate_report(df)
```

```python
#The above code defines a function to filter reviews based on rating, helpfulness, and keyword presence.
#It starts with the original DataFrame and applies the specified filters if criteria are provided.
#An example demonstrates filtering for reviews with a rating of 5, at least 10 helpful votes, and containing the keyword 'great'.
#The filtered_reviews.head() function shows the first few rows of the filtered results.
```

```python
#Review filtering feature
import pandas as pd

# Function to filter reviews based on criteria
def filter_reviews(df, rating=None, helpfulness=None, keyword=None):
    filtered_df = df
    if rating is not None:
        filtered_df = filtered_df[filtered_df['reviews.rating'] == rating]
    if helpfulness is not None:
        filtered_df = filtered_df[filtered_df['reviews.numHelpful'] >= helpfulness]
    if keyword is not None:
        filtered_df = filtered_df[filtered_df['reviews.text'].str.contains(keyword, case=False)]
    return filtered_df

# Example usage
filtered_reviews = filter_reviews(df, rating=5, helpfulness=10, keyword='great')
filtered_reviews.head()
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd

# Load your dataset
df = pd.read_csv('amazon.csv')  # Adjust the path accordingly

# Filter function
def filter_reviews(df, rating=None, helpfulness=None, keyword=None):
    filtered_df = df
    if rating is not None:
        filtered_df = filtered_df[filtered_df['reviews.rating'] == rating]
    if helpfulness is not None:
        filtered_df = filtered_df[filtered_df['reviews.numHelpful'] >= helpfulness]
    if keyword is not None:
        filtered_df = filtered_df[filtered_df['reviews.text'].str.contains(keyword, case=False)]
    return filtered_df

# Streamlit app for review filtering
st.title('Amazon Reviews Filter')

rating = st.slider('Select Rating', 1, 5, value=3)
helpfulness = st.slider('Select Minimum Helpfulness', 0, 100, value=10)
keyword = st.text_input('Enter Keyword')

filtered_reviews = filter_reviews(df, rating=rating, helpfulness=helpfulness, keyword=keyword)
st.write(filtered_reviews)
```

It loads the dataset from a CSV file and uses sliders and a text input to allow users to specify filtering criteria. The filtered results are displayed in the app. The application provides an interactive way to explore and analyze review data.
